```python
#Difference between Series, DataFrame and Panel in Pandas 
Series can only contain single list with index, 
whereas dataframe can be made of more than one series. 
we can say that a dataframe is a collection of series that can be used to analyse the data.
```

```python
#Write a program to concat two DataFrames.

# Creating first dataframe
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2'],
                    'C': ['C0', 'C1', 'C2'],},
                    index = [0, 1, 2])
  
# Creating second dataframe
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6'],
                    'B': ['B4', 'B5', 'B6'],
                    'C': ['C4', 'C5', 'C6'],},
                    index = [4, 5, 6])
  

# Concatenating the dataframes
pd.concat([df1, df2])
  
```

```python
#Write a program to merge two DataFrames. 
# Creating first dataframe
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2'],
                    'C': ['C0', 'C1', 'C2'],},
                    index = [0, 1, 2])
  
# Creating second dataframe
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6'],
                    'B': ['B4', 'B5', 'B6'],
                    'C': ['C4', 'C5', 'C6'],},
                    index = [4, 5, 6])
  

# merging two the dataframes
pd.concat([df1, df2])
  
```

```python
#Write a program to handle missing data from DataFrame. 
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', np.nan, 'B2'],
                    'C': ['C0', 'C1', 'C2'],},
                    index = [0, 1, 2])
  
# Creating second dataframe
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6'],
                    'B': ['B4', 'B5', 'B6'],
                    'C': ['C4', np.nan, 'C6'],},
                    index = [4, 5, 6])
df3=pd.concat([df1, df2])
  
```
